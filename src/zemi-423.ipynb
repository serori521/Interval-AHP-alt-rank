## 区間AHPにおける順位変化点検出アルゴリズム実装

### 1. miniMax regret手法

miniMax regret手法では、各代替案の最大リグレットを最小化することで、不確実な区間重要度のもとでの順位付けを行う。式に表すと：

\begin{align*}
    \text{maximize} \quad & R_j^o = \sum_{i \in N} w_i(u_i(O_j) - u_i(O_o)) \\
    \text{s.t.} \quad & \sum_{i \in N} w_i = 1 \\
\end{align*}

ここで、$R_j^o$は代替案$O_j$の代替案$O_o$に対するリグレット値であり、他の代替案も比べてその最大リグレットを求める。以下では一つの代替案の最大リグレットがどのように変化するか、その変化点を求めるアルゴリズムを示す。

### 2. 一つの代替案の最大リグレットを与える代替案の変化点検出アルゴリズム

区間AHPでは、tの値によって各代替案間の順位が変化する。このtの値（順位変化点）を効率的に検出するためのアルゴリズム。

#### 3.1 傾きが変わらない区間$[t^0, t^1]$内でのアルゴリズム

1. **区間$[t^0, t^1]$内での各代替案$o_o$に対する最大リグレット**
   - 各代替案$o_q$に対して、$r(o_q : o_o, t)$を$o_o$に対する$o_q$のリグレットとして定義

2. **最大リグレットを与える候補集合の初期化**
   - $o_o$との効用差から、区間内のいずれかの点で最大リグレットを与える代替案の集合$\hat{O}_o$と、$[t^0, t^1]$内での$o_o$のリグレットの最大値$r^M$と最小値$r^m$を計算

3. **初期値の設定**
   - $r(o_p : o_o, t^0)$を$t^0$における$o_o$の最大リグレット（第一候補）とする
   - $r(o_u : o_o, t^0)$を二番目のリグレットとし、初期状態として$\hat{O}_o = \{o_p\}$とし、各候補$o_q$に対して$t_o(o_q) = t^0$と初期化

4. **区間上端での候補抽出**
   - 区間上端$t^1$において、$r(o_q : o_o, t^1) > r(o_p : o_o, t^1)$となるすべての$o_q$により「1番になりうる代替案」の集合$O$を定める

5. **各候補との交点の算出**
   - 各$o_q \in O$について$r(o_q : o_o, t) = r(o_p : o_o, t)$となる$t$を求め、これを$t_o(o_q)$とする

6. **不要候補の除去**
   - $t_o(o_s) < t_o(o_q)$かつ$r(o_s : o_o, t^1) > r(o_q : o_o, t^1)$となる$o_s \in O$が存在すれば、$O$から$o_q$を除く

7. **更新ステップによる区間再設定**
   - $O$内の候補について、$r(o_s : o_o, t_o(o_u)) > r(o_u : o_o, t_o(o_u))$となる$o_s \in O$が存在すれば、$t^0$を$t_o(o_s)$に更新し、$o_q = o_s$としてステップ①に戻る
   - なければ、$t^0 = t_o(o_u)$および$o_q = o_u$として再びステップ①へ

8. **最終的な候補集合とリグレットの決定**
   - 最終的に、$\hat{O}_o = O \cup \{o_p\}$とし、リグレットの最大値・最小値を決定
